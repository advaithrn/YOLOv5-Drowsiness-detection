```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret,frame = cap.read()
    
    #make detections
    results = model(frame)
    print(results)
    cv2.imshow('YOLO',np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
